```python
import numpy as np
import pandas as pd
import os
from google.colab import drive
drive.mount('/content/drive')
df_Doc = pd.read_csv('/content/drive/My Drive/Data_viz_2024/Data_Docter_ศ.นพ.ชวลิต.csv')
```

```python
# คำนวณยอดไลค์เฉลี่ยต่อการตอบกลับ (เลี่ยงการหารด้วย 0)
df_Doc['ไลค์ต่อการตอบกลับ'] = df_Doc['ยอดไลค์'] / df_Doc['จำนวนการตอบกลับ'].replace(0, 1)

# หาผู้ที่มีไลค์ต่อการตอบกลับสูงสุด
max_likes_per_response = df_Doc.loc[df_Doc['ไลค์ต่อการตอบกลับ'].idxmax()]

max_likes_per_response[['name', 'ไลค์ต่อการตอบกลับ','comments']]
```

```python
# ค้นหาผู้ที่มียอดไลค์สูงแต่จำนวนการตอบกลับต่ำ
df_low_responses_high_likes = df_Doc[(df_Doc['ยอดไลค์'] > 100) & (df_Doc['จำนวนการตอบกลับ'] <= 3)]

df_low_responses_high_likes[['name', 'ยอดไลค์', 'จำนวนการตอบกลับ','comments']]
```
