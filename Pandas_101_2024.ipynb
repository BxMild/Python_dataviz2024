````
# prompt: วนลูปว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง
# มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
# คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด

# วนลูปว่ามีนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง
count_mom_dad = 0
mom_dad_titles = []
for index, row in folk_df.iterrows():
  if 'แม่' in row['title'] or 'พ่อ' in row['title']:
    count_mom_dad += 1
    mom_dad_titles.append(row['title'])

print(f"จำนวนนิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง: {count_mom_dad}")
print("รายชื่อนิทาน:")
for title in mom_dad_titles:
  print(title)

# มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
count_life = 0
count_local = 0
for index, row in folk_df.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']):
    if row['type'] == 'นิทานชีวิต':
      count_life += 1
    elif row['type'] == 'นิทานประจำถิ่น':
      count_local += 1

if count_life > count_local:
  print("มีคำว่า แม่ หรือ พ่อ ในนิทานชีวิตมากกว่า")
elif count_local > count_life:
  print("มีคำว่า แม่ หรือ พ่อ ในนิทานประจำถิ่นมากกว่า")
else:
  print("มีคำว่า แม่ หรือ พ่อ ในนิทานชีวิตและนิทานประจำถิ่นเท่ากัน")

# คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด
creator_counts = {}
for index, row in folk_df.iterrows():
  if ('แม่' in row['title'] or 'พ่อ' in row['title']):
    creator = row['creator']
    if creator not in creator_counts:
      creator_counts[creator] = 0
    creator_counts[creator] += 1

max_count = 0
most_frequent_creator = ""
for creator, count in creator_counts.items():
  if count > max_count:
    max_count = count
    most_frequent_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุดคือ: {most_frequent_creator}")
````

บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่าเป็นผู้กำกับชั้นนำ และชั้นนำยังไง

(rating Q3 ขึ้นไป)

````
# prompt: บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่าเป็นผู้กำกับชั้นนำ และชั้นนำยังไง
# (ดูจาก rating ที่มากกว่า 75% และ Votesที่มากกว่า 75%) พร้อมแสดงข้อมูลของผู้กำกับ และตัวอย่างหนัง 3 เรื่องที่ rating และ votes มากสุด ของผู้กำกับแต่ละคน)

# อ่านข้อมูลจากไฟล์ CSV
moviedf = pd.read_csv('/content/drive/My Drive/Data_viz_2024/IMDb Movies India.csv', encoding='latin-1')

# Clean the 'Votes' column to remove non-numeric characters and convert to float
moviedf['Votes'] = moviedf['Votes'].str.replace(',', '').str.replace('$', '').str.replace('M', '').astype(float)

# คำนวณ rating เป็น %
moviedf['Rating_percent'] = moviedf['Rating'] * 10

# เลือกผู้กำกับที่มี rating มากกว่า 75% หรือ Votes มากกว่า 75 percentile
top_directors = moviedf[(moviedf['Rating_percent'] > 75) | (moviedf['Votes'] > moviedf['Votes'].quantile(0.75))]['Director'].value_counts()

# แสดงข้อมูลผู้กำกับชั้นนำ 3 คน
for director, count in top_directors.head(3).items(): # Iterate over director and count pairs
  print(f"{director} :{count} movies")
  # Use the correct DataFrame variable name 'moviedf'
  director_movies = moviedf[moviedf['Director'] == director].sort_values(by=['Rating', 'Votes'], ascending=False)
  print(f"* Average rating: {director_movies['Rating'].mean():.2f}")
  print(f"* Average votes: {director_movies['Votes'].mean():,.0f}")
  print("   ตัวอย่างหนัง 3 เรื่อง:")
  for i in range(min(3, director_movies.shape[0])):
    movie = director_movies.iloc[i]
    print(f"   - {movie['Name']} ({movie['Year']}), Rating: {movie['Rating']}, Votes: {movie['Votes']}")
  print("\n")
````
