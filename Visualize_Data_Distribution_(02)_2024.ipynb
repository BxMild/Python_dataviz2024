# การแสดงผลการกระจายของข้อมูล

````
import pandas as pd
from matplotlib import pyplot as plt
````

````
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df.head()
````

## Scatter 2D

## scatter - default

````
plt.scatter(df['PetalLength'],df['PetalWidth'])
````

### color

````
plt.scatter(df['PetalLength'],df['PetalWidth'],c='r')
````

## ใส่สีให้แต่ละจุด

````
set(df['Name']) #ดูว่ามีค่าอะไรบ้าง
````

````
#ใส่เลขแทนชื่อเข้าไป
df2 = df.replace({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica':2})
df2
````

````
plt.scatter(df['PetalLength'],
            df['PetalWidth'],
            c=df2['Name'])
````

### alpha กำหนดความโปร่งแสง

````
#จะมองเห็นความหนาของของมูลจากความโปร่งใส่ ทีทับกัน
plt.scatter(df['PetalLength'],df['PetalWidth'],c=df2['Name'],alpha=0.5)
````

## ใช้ขนาดของ marker แสดงค่าของ feature

````
plt.scatter(df['PetalLength'],
            df['SepalLength'],
            s=30*df['PetalWidth'],
            c=df2['Name'],
            alpha=0.5)
````

## ตัวแปร Marker ใช้ในการเปลี่ยนหน้าตาของ marker

````
plt.scatter(df['PetalLength'],df['PetalWidth'],s=100,c=df2['Name'],alpha=0.5,marker="*")
````

## การ plot กราฟหลายๆอันซ้อนกัน

````
plt.scatter(df['PetalLength'][:50],df['PetalWidth'][:50],s=100,c='m',alpha=0.5,marker='$S$')
plt.scatter(df['PetalLength'][50:100],df['PetalWidth'][50:100],s=100,c='c',alpha=0.5,marker='$E$')
plt.scatter(df['PetalLength'][100:],df['PetalWidth'][100:],s=100,c='r',alpha=0.5,marker='$I$')
````

###Label-Legend

```plt.plot()``` input  ตัวที่ 3 คือ ตัวกำหนดหน้าตาของ marker

'หน้าตาmarker'-'ลักษณะของเส้นเชื่อมจุด'-'สี'

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],
         df['PetalWidth'][:50],
         'o-g',#'รูปแบบ เส้น สี'
         alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'o:b',alpha=0.5)
````

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5,label='setosa')
plt.plot(df['PetalLength'][50:100],df['PetalWidth'][50:100],'xc',alpha=0.5,label='versicolor')
plt.plot(df['PetalLength'][100:],df['PetalWidth'][100:],'*m',alpha=0.5,label='virginica')
plt.legend(loc='upper left')
````

### การใส่ชื่อกราฟ และชื่อแกน

````
plt.plot(df['PetalLength'][:50],df['PetalWidth'][:50],'or',alpha=0.5,label='setosa')
plt.plot(df['PetalLength'][50:100],df['PetalWidth'][50:100],'xc',alpha=0.5,label='versicolor')
plt.plot(df['PetalLength'][100:],df['PetalWidth'][100:],'*m',alpha=0.5,label='virginica')
plt.legend(loc='lower right')
plt.title('Iris flower data distribution') ##ชื่อกราฟ
plt.xlabel('PetalLength') ##ชื่อแกน X
plt.ylabel('PetalWidth'); ##ชื่อแกน y
````

## 3D scatter

````
ax = plt.axes(projection ="3d")
ax.scatter3D(df['PetalLength'],df['PetalWidth'],df['SepalWidth'],c=df2['Name'])
````

## Bubble Chart

Scatter plot ที่ใช้ขนาดของ marker ในการสื่อปริมาณของข้อมูล

````
plt.scatter(df['PetalLength'][:50],df['SepalLength'][:50],s=180*df['PetalWidth'][100:],c='r')
````

````
import numpy as np
import pandas as pd
import os
from google.colab import drive
drive.mount('/content/drive')
````

````
df_movies = pd.read_csv('/content/drive/My Drive/Data_viz_2024/IMDb Movies India.csv', encoding='latin-1') # try different encodings like 'latin-1' or 'ISO-8859-1' if this doesn't work
````

````
df_movies.head()
````

````
# แปลงคอลัมน์ 'Rating' และ 'Votes' ให้เป็นตัวเลข
df_movies['Rating'] = pd.to_numeric(df_movies['Rating'], errors='coerce')
df_movies['Votes'] = pd.to_numeric(df_movies['Votes'], errors='coerce')

# ลบแถวที่มีค่า NaN ในคอลัมน์ 'Rating' หรือ 'Votes'
df_clean = df_movies.dropna(subset=['Rating', 'Votes'])

df_clean.head()
````

````
df_movies = df_movies.dropna()
````

````
df_movies['Year'] = df_movies['Year'].str.replace('(', '').str.replace(')', '').astype(int)
````

````
df_movies.head()
````

````
# Use 'Rating' for the histogram and optionally specify bins
plt.hist(df_clean['Rating'], bins=10)

plt.legend(loc='lower right')
plt.title('Rating vs Votes')
plt.xlabel('Rating')
plt.ylabel('Votes');
````

````
rating_per_year = df_movies.groupby('Year')['Rating'].mean().reset_index(name='Average Rating')
````

````
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(rating_per_year['Year'], rating_per_year['Average Rating'],'o-')

# เพิ่มรายละเอียดกราฟ
plt.title('Average Rating of Movies by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)  # หมุนป้ายปีเพื่อให้ดูง่ายขึ้น
plt.grid(True)

# แสดงกราฟ
plt.show()
````
