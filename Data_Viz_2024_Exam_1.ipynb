```python
import os
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
path_to_test = '/content/drive/My Drive/Data_viz_2024/test-TidyData_2024.xlsx'
tast_df = pd.read_excel(path_to_test)
tast_df
```

```python
#1. ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019
tast_df['Year'] = tast_df['month'].apply(lambda x: int(x.split('_')[-1]))

tast_df_2018 = tast_df[tast_df['Year'] == 2018]
tast_df_2019 = tast_df[tast_df['Year'] == 2019]

mean_chlo_a_2018 = tast_df_2018.groupby('station')['Chlo_a'].mean()
mean_chlo_a_2019 = tast_df_2019.groupby('station')['Chlo_a'].mean()

mean_chlo_a = pd.DataFrame({'Mean Chlo_a 2018': mean_chlo_a_2018, 'Mean Chlo_a 2019': mean_chlo_a_2019})
mean_chlo_a
```

```python
#2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน
tast_df['month_N'] = tast_df['month'].apply(lambda x:x.split('_')[0])

mean_chlo_b_by_month_station = tast_df.groupby(['month_N', 'station'])['Chlo_b'].mean().unstack()
mean_chlo_b_by_month_station
```
